# preprocess

```python
import pandas as pd
```

```python
df = pd.read_json("data/SQuAD1.0/train-v1.0.json")
```

```python
data_list = pd.DataFrame({'context': [],
                   'answer': [],
                   'question': []})

for data in df["data"]:
    for paragraph in data["paragraphs"]:
        for qa in paragraph["qas"]:
            data_list = data_list.append({'context': [paragraph["context"]],
                   'answer': [qa["answers"][0]["text"]],
                   'question': [qa["question"]]}, ignore_index=True)
```

```python
data_list.to_json("data/SQuAD1.0/train-v1.0-formatted.json")
```
